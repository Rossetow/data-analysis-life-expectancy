```python
# Dependecies

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt

from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import Dense, LSTM

plt.style.use('fivethirtyeight')
```

```python
# Get the dataframe

df = pd.read_csv('Life_Expectancy_Data.csv')

df.head()
```

```python
df.info()
```

```python
# Checking null data

df.isnull().sum()
```

```python
df = df.dropna(axis = 0)
```

```python
df.isnull().sum()
```

```python
df = df[((df.Population != 'Unknown'))]

df['Population'] = pd.to_numeric(df['Population'])
```

```python
df.head(55)
```

```python
df.info()
```

```python
# Get the useful data from the initial dataset

clean_data = df.filter(['Year', 'Continent', 'Country', 'Expectancy'])
```

```python
clean_data.head()
```

```python
# Average life expectancy by continent

x = sorted(df['Continent'].unique())

y = df.groupby(['Continent'])['Expectancy'].mean()

plt.figure(figsize = (16, 8))

plt.bar(x, y)

plt.show()
```

```python
# Filter method

def filter(continent):
    return clean_data['Continent'] == continent

# Average life expentacy by country by continent

countries_by_region = []
for i in range(len(clean_data['Continent'].unique())):
    countries_by_region.append(clean_data.where(filter(clean_data['Continent'].unique()[i])).dropna())

for i in range(len(countries_by_region)):
    x = sorted(countries_by_region[i]['Country'].unique())

    y = countries_by_region[i].groupby(['Country'])['Expectancy'].mean()

    plt.figure(figsize = (8, len(y)*.5))

    plt.title(clean_data['Continent'].unique()[i])

    plt.barh(y=x, width=y)

    plt.xlabel('Life expectancy')

    plt.show()
```

```python
# Highest and lowest life expentacy

# Last record of each country (2015)

all_countries_last_ten_years = clean_data.groupby('Country').head(1).reset_index(drop=True)

# Get the highest and lowest

x_highest, y_highest = all_countries_last_ten_years.sort_values('Expectancy').tail(1)['Country'], all_countries_last_ten_years['Expectancy'].max()

x_lowest, y_lowest = all_countries_last_ten_years.sort_values('Expectancy').head(1)['Country'],all_countries_last_ten_years['Expectancy'].min()
```

```python
# Difference between the highest and lowest in a graphic

plt.figure(figsize = (8, len(y)*.5))

plt.title(clean_data['Continent'].unique()[i])

plt.barh(y=(x_highest.array[0], x_lowest.array[0]), width=(y_highest, y_lowest))

plt.xlabel('Life expectancy', fontsize = 15)

plt.show()
```
