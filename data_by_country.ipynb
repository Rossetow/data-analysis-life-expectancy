```python
# Dependecies

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt

plt.style.use('fivethirtyeight')
```

```python
# Get the dataframe

# Data source: https://gist.github.com/aishwarya8615/89d9f36fc014dea62487f7347864d16a#file-life_expectancy_data-csv

df = pd.read_csv('Life_Expectancy_Data.csv')

# Dropping rows that contain null data

df = df.dropna(axis = 0)

# Dropping rows in the column "Population" in which the content is
# equal to 'Unknown' to avoid conflicts when converting the data from
# Object to numeric

df = df[((df.Population != 'Unknown'))]

# Converting Population from Object to float64

df['Population'] = pd.to_numeric(df['Population'])

df.head()
```

```python
country = ''

print("Of what country do you wanto to see informatios?" )
print("Please note that the program is case sensitive: ")

country = input()
```

```python
# Function for filtering the DataFrame

def filter_by_country(country):
    return df['Country'] == country

df_country = df.where(filter_by_country(country)).dropna()

df_country
```

```python
# Get life expectancy for a specific country

# Function for getting a DataFrame with only the information of the selected country

# X as the years and Y as the expectancy
x, y = (df_country['Year'], df_country['Expectancy'])

if(len(y) == 0):
    print("Error: " + country + " not found!")
else:
    #Graph

    plt.figure(figsize=(16, 8))
    
    plt.title("Life expentacy through the years in " + country)

    plt.xlabel("Year")

    plt.ylabel("Age")
    
    plt.plot(x.array, y.array)
    
    plt.show()
```

```python
# Your country compared with the global average

if(len(y) == 0):
    print("Error: " + country + " not found!")
else:
    all_countries_last_year_average = df.groupby('Country').head(1).reset_index(drop=True)['Expectancy'].mean()
    
    # Get your country's latest entry
    
    most_recent_expectancy = df_country.sort_values('Year').tail(1)['Expectancy'].array[0]
    
    # List of continents sorted
    
    x = []
    
    x.append("World")
    x.append(country)
    
    # Average life expectancy grouped by continent
    
    y = []
    
    y.append(all_countries_last_year_average)
    
    y.append(most_recent_expectancy)
    
    # Graph
    
    plt.figure(figsize = (16, len(y)*.5))
    
    plt.title(country + " compared with the global average")
    
    plt.barh(y = x, width = y)
    
    plt.show()
```

```python
# Your country vs the hoghest and the lowest life expectancies

if(len(y) == 0):
    print("Error: " + country + " not found!")
else:
    # Last record of each country (2015)
    
    all_countries_last_year = df.groupby('Country').head(1).reset_index(drop=True)
    
    # Get the highest and lowest
    
    x_highest, y_highest = all_countries_last_year.sort_values('Expectancy').tail(1)['Country'], all_countries_last_year['Expectancy'].max()
    
    x_lowest, y_lowest = all_countries_last_year.sort_values('Expectancy').head(1)['Country'],all_countries_last_year['Expectancy'].min()
    
    # Compile information for the graph
    
    y_graph = [x_lowest.array[0], country, x_highest.array[0]]
    x_graph = [y_lowest, most_recent_expectancy, y_highest]
    
    # Show graph
    
    plt.figure(figsize = (16, len(y_graph)*.5))
    
    plt.title(country + " compared with highest and lowest life expectancies")
    
    plt.barh(y = y_graph, width = x_graph)
    
    plt.xlabel('Life expectancy', fontsize = 15)
    
    plt.show()
```

```python
# Life expectancy in your country in comparison with the continent

if(len(y) == 0):
    print("Error: " + country + " not found!")
else:
    # Average life expectancy by continent
    
    # List of continents sorted
    
    x_graph = []
    
    x_graph.append(df_country.head(1)['Continent'].array[0])
    x_graph.append(country)
    
    # Filter continent for the next graph
    
    def filter_continent():
        return df['Continent'] == x[0]
    
    # Average life expectancy grouped by continent
    
    countries_in_the_same_continent = df.where(filter_continent()).dropna()
    
    y_graph = []
    
    y_graph.append(countries_in_the_same_continent['Expectancy'].mean())
    
    y_graph.append(most_recent_expectancy)
    
    # Graph
    
    plt.figure(figsize = (16, len(y)*.5))
    
    plt.title(country + " compared with the continent")
    
    plt.barh(y = x, width = y)
    
    plt.show()
```
